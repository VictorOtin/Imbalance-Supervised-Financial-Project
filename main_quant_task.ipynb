```python
def make_X_y(df_test_transformed):
    target_class = 'classlabel_yes.'
    X_test = df_test_transformed.copy()
    X_test = df_test_transformed.drop(columns = [target_class])
    y_test = df_test_transformed[target_class]
    
    return X_test, y_test

def split_dataframe(df, chunk_size = 10): 
    chunks = list()
    num_chunks = len(df) // chunk_size + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunk_size:(i+1)*chunk_size])
    return chunks


def eval_set_mean_std(df_test_transformed,clf,kfolds):
    
    flag = 0
    scores_roc = []
    scores_f1 = []
    splits = split_dataframe(df_test_transformed,chunk_size = int(len(df_test_transformed) / kfolds))
    for df_split in splits:
        if len(df_split) == 0: continue
        X_test,y_test = make_X_y(df_split)
        
        try:
            probs = clf.predict_proba(X_test)[:, 1]
            y_pred = clf.predict(X_test)
            scores_roc.append(roc_auc_score(y_test, probs))
            flag = 1
        except AttributeError:
            pass # does nothing
            
        y_pred = clf.predict(X_test)
        scores_f1.append(f1_score(y_test, y_pred, average='macro'))
        
    if flag == 1: print('Mean_roc_scores_TEST:', mean(scores_roc), 'Std_roc_scores_TEST:',std(scores_roc))
    print('Mean_f1_scores_TEST', mean(scores_f1), 'Std_f1_scores_TEST:',std(scores_f1))
    return X_test,y_test
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.model_selection import cross_validate
import shap



def modelling(X_train, y_train,df_test_transformed):
    h = .02  # step size in the mesh
    kfolds = 5

    names = ["Nearest Neighbors", "Linear SVM", "RBF SVM",
             "Decision Tree", "Random Forest","RF class weight","RF bagging imbalanced", "Neural Net", "AdaBoost",
             "Naive Bayes","Logistic Regression"] 

    tree = RandomForestClassifier(n_estimators=100,class_weight='balanced',max_depth = 3)


    classifiers = [
        KNeighborsClassifier(3),
        SVC(kernel="linear", C=0.025),
        SVC(gamma=2, C=1),
        DecisionTreeClassifier(max_depth=5),
        RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
        RandomForestClassifier(n_estimators=100,class_weight='balanced',max_depth = 3),
        BalancedBaggingClassifier(base_estimator=tree,random_state=0),
        MLPClassifier(alpha=1, max_iter=1000),
        AdaBoostClassifier(),
        GaussianNB(),
        LogisticRegression(class_weight = 'balanced')]

    # iterate over classifiers
    models_dict = {}
    for name, clf in zip(names, classifiers):
        print()
        print('-------------------', name, '-------------------')
        clf.fit(X_train, y_train)

        scores_train = cross_validate(clf,
                                X_train, y_train, cv=kfolds,
                                scoring=('f1','roc_auc'))
        print('f1-mean:',scores_train['test_f1'].mean(),
              'f1-std:',scores_train['test_f1'].std(),
              'roc-auc-mean:',scores_train['test_roc_auc'].mean(),
              'roc-auc-std:',scores_train['test_roc_auc'].std())

        clf.fit(X_train, y_train)


        X_test,y_test = eval_set_mean_std(df_test_transformed,clf,kfolds = 5)


        models_dict[name] = clf
        
    return models_dict


models_dict= modelling(X_train, y_train,df_test_transformed)
```

```python
shap.initjs()

def shap_knn(models_dict,X_test):
    explainer = shap.KernelExplainer(models_dict["Nearest Neighbors"].predict_proba, X_test)
    shap_values_ = explainer.shap_values(X_test)
    #shap.force_plot(explainer.expected_value[0], shap_values_[0], X_test)
    shap.summary_plot(shap_values_, X_test)
    plt.show()
    
def shap_rf(models_dict,X_train,X_test):
    
    explainer = shap.TreeExplainer(models_dict["Random Forest"])
    shap_values = explainer.shap_values(X_train)
    shap.summary_plot(shap_values, X_train)
    plt.show()
    
    shap_values = explainer.shap_values(X_test)
    shap.summary_plot(shap_values, X_test)
    plt.show()

shap_rf(models_dict,X_train,X_test)
```

## With PCA

```python
from sklearn.decomposition import PCA
import pickle

def do_PCA(X_train,X_test,n_comp = 7):
    pca = PCA(n_components = n_comp)
    pca.fit(X_train)
    print('Explained variance:',pca.explained_variance_ratio_.sum())

    return pca.transform(X_train),pca.transform(X_test)


X_train_pca, X_test_pca = do_PCA(X_train = X_train, X_test = X_test)
```

```python
df_test_pca = pd.DataFrame(X_test_pca)
df_test_pca['classlabel_yes.'] = y_test

X_test_pca = pd.DataFrame(X_test_pca)

X_train_pca = pd.DataFrame(X_train_pca)
```

```python
df_test_pca.shape, X_train_pca.shape
```

```python
models_dict_pca = modelling(X_train = X_train_pca, y_train = y_train ,df_test_transformed = df_test_pca)
```

Much better! RF bagging for imbalanced get good results and it seems it overcome the imbalance difficulty if we compare it with the random forest baseline

```python
shap_rf(models_dict_pca,X_train_pca,X_test_pca)
```

Here we don't have to do cluster feature importance since the eigenvectors are already independent.

There is a correspondence between the important features in train and test, it is something good. The last features are not the same in training and test and it might have shrinked the performance in validation set. It could be improved even further addresing
